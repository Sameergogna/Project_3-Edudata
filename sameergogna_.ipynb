# Q5

```python
X=edudata1.drop(["ParentschoolSatisfaction_Good"], axis=1)
y=edudata1["ParentschoolSatisfaction_Good"]
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train , y_test=train_test_split(X,y, test_size=0.25, random_state=88)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log0=LogisticRegression()
```

```python
log0.fit(X_train, y_train) 
```

```python
print("Train Accuracy :", log0.score(X_train, y_train))  
```

```python
print("Test Accuracy :", log0.score(X_test, y_test))
```

```python
pred_train=log0.predict(X_train) 
```

```python
y_train
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_train, pred_train)
```

```python
print(metrics.classification_report(y_train, pred_train)) 
```

```python
pred_test=log0.predict(X_test) 
```

```python
metrics.confusion_matrix(y_test, pred_test)
```

```python
print(metrics.classification_report(y_test, pred_test)) 
```
